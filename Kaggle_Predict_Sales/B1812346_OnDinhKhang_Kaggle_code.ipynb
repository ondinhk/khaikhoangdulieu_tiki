# Chia tập dữ liệu

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
sales_train.head()
```

# Định dạng lại tập train

```python
# Bỏ cột date, date_block_num, item_price
sales_train = sales_train.drop(columns=['date','date_block_num','item_price'])
sales_train.head()
```

```python
y = sales_train['item_cnt_day']
x = sales_train.drop(columns=['item_cnt_day'])
# Chia tập dữ liệu
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size = 0.3, random_state = 60,shuffle=True)
```

```python
# X_train
X_train.head()
```

```python
# Y_train
Y_train.head()
```

# Kích thước tập train

```python
print("Kích thước tập train: ", X_train.shape)
print("Kích thước nhãn train:", Y_train.shape)
```

# Kích thước tập test

```python
print("Kích thước tập test: ", X_test.shape)
print("Kích thước nhãn test:", Y_test.shape)
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
#Define The Model
dt_regressor = DecisionTreeRegressor()
# Fitting
dt_regressor.fit(X_train,Y_train)
# Predict 
dt_pred = dt_regressor.predict(X_test)
# RMSE 
rmse = mean_squared_error(Y_test,dt_pred,squared=False)
rmse
# 1.9863306628505226
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
#Define The Model
ln_regressor = LinearRegression()
# Fitting
ln_regressor.fit(X_train,Y_train)
# Predict
ln_pred = ln_regressor.predict(X_test)
# MRSE
rmse = mean_squared_error(Y_test,ln_pred,squared=False)
rmse
# 2.3860295729858696
```

# KNeighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
#Define The Model
knn = KNeighborsClassifier(n_neighbors=3)
# Fitting
knn.fit(X_train, Y_train)
# MRSE
knn_pred = knn.predict(X_test)
rmse = mean_squared_error(Y_test,knn_pred,squared=False)
rmse
# 2.323155481969899
```
